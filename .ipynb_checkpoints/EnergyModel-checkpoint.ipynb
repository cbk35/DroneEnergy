```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Load data from Excel file
data = pd.read_excel('testflightdata3.xlsx')

# Define the feature columns
feature_cols = ['Speed (m/s)','Weight (g)','Equivalent Wind (m/s)'] #'Altitude (m)','Equivalent Wind (m/s)']
data_filtered = data.loc[:,['Speed (m/s)','Weight (g)','Altitude (m)','Wind (m/s)','Calc. Power (W)']]

# Split data into features and target
X = data.loc[:,feature_cols]
y = data.loc[:,['Calc. Power (W)']]

# Split data into training and testing sets
train_data, test_data = train_test_split(data_filtered, test_size=0.3, random_state=5)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
display(train_data)
```

```python
# Normalizer and autoencoder
train_target = y_train
test_target = y_test

# Normalize data
scaler = StandardScaler()
train_data = scaler.fit_transform(train_data)
test_data = scaler.transform(test_data)

# Create autoencoder
input_dim = train_data.shape[1]
encoding_dim = 32

# Add nodes to autoencoder
autoencoder = Sequential()
autoencoder.add(Dense(256, activation='relu', input_shape=(input_dim,)))
autoencoder.add(Dense(128, activation='relu'))
autoencoder.add(Dense(encoding_dim, activation='relu'))
autoencoder.add(Dense(128, activation='relu'))
autoencoder.add(Dense(256, activation='relu'))
autoencoder.add(Dense(input_dim, activation='linear'))

# Compile autoencoder
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# Train autoencoder
autoencoder.fit(train_data, train_data, epochs=100, batch_size=32, validation_data=(test_data, test_data))

# Obtain encoded data
encoded_train_data = autoencoder.predict(train_data)
encoded_test_data = autoencoder.predict(test_data)

display(train_data)
display(encoded_train_data)
display(test_data)
display(encoded_test_data)
```

```python
# Create an XGBoost regressor
xgb_model = xgb.XGBRegressor(objective='reg:squarederror')

# Define the hyperparameters to tune
param_grid = {'learning_rate': [0.05, 0.1, 0.2],
              'max_depth': [3, 5, 7],
              'n_estimators': [50, 100, 200],
              'subsample': [0.5, 0.75, 1.0],
              'colsample_bytree': [0.5, 0.75, 1.0],
              'gamma': [0, 0.1, 0.5]}

# Perform grid search to find the best hyperparameters
grid_search = GridSearchCV(xgb_model, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)
#grid_search.fit(encoded_train_data, train_target) # Enable wehen using autoencoder

# Print the best hyperparameters and the corresponding score
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Score: ", np.sqrt(-grid_search.best_score_))

# Use the best model to make predictions on the test set
y_pred = grid_search.predict(X_test)

# Evaluate model on test set using encoded test data
#y_pred = grid_search.predict(encoded_test_data) # Enable wehen using autoencoder
y_test=test_target
```

```python
# Calculate mean squared error
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)

# Calculate root mean squared error
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Root Mean Squared Error:", rmse)

# Calculate mean absolute error
mae = mean_absolute_error(y_test, y_pred)
print("Mean Absolute Error:", mae)

# Calculate R-squared
r2 = r2_score(y_test, y_pred)
print("R-squared:", r2)

# Plot predicted values vs actual values
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Actual vs Predicted Power Consumption")
plt.ylim([100, 400])
plt.xlim([100, 400])
plt.grid(True)
plt.show()
print(y_test)
print(y_pred)
```

```python
# Plot multiline chart
weights = [1102, 1418, 1734]
colors = ['b', 'g','r']
plt.figure(figsize=(10, 6))

for i, weight in enumerate(weights):
    X_sub = X_test[X_test['Weight (g)'] == weight]
    y_sub = y_test[X_test['Weight (g)'] == weight]
    y_pred_sub = y_pred[X_test['Weight (g)'] == weight]
    speed_vals = X_sub['Speed (m/s)'].values

    # Sorting
    z = list(zip(speed_vals, y_sub.values, y_pred_sub))
    
    # Sort the list of tuples by the first element of each tuple (i.e., the integer from x) in ascending order
    z_sorted = sorted(z, key=lambda t: t[0])
    
    # Extract the sorted integers from x and y as separate lists
    speed_vals_sorted = [t[0] for t in z_sorted]
    y_sub_values_sorted = [t[1] for t in z_sorted]
    y_pred_sub_sorted = [t[2] for t in z_sorted]  
    
    plt.plot(speed_vals_sorted, y_sub_values_sorted, linestyle='--', marker='o', color=colors[i], label=f'Weight={weight} g Test Data')
    plt.plot(speed_vals_sorted, y_pred_sub_sorted, linestyle='-', marker='x', color=colors[i], label=f'Weight={weight} g Predicted')

plt.title('XGBoost Regressor: Power vs Speed per Different Weights')
plt.xlabel('Speed (m/s)')
plt.ylabel('Power (W)')
plt.legend()
plt.show()
```

```python
# Create a linear regression object and fit the data
lr = LinearRegression().fit(X, y)

# Print the coefficients of the model
print(lr.coef_)

# Evaluate the model on the training data
y_pred = lr.predict(X)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"MSE: {mse:.3f}")
print(f"RMSE: {rmse:.3f}")
print(f"MAE: {mae:.3f}")
print(f"R-squared: {r2:.3f}")
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_selection import SelectKBest, f_regression

# Create polynomial features up to degree 3
poly = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly.fit_transform(X)

# Select the top 10 significant features
selector = SelectKBest(f_regression, k=10)
selector.fit(X_poly, y)
X_new = selector.transform(X_poly)

# Fit linear regression with the selected features
reg = LinearRegression()
reg.fit(X_new, y.values.ravel())
coeffs = reg.coef_
#np.set_printoptions(precision=8)

# Print the formula
formula = "y = {:.4f}".format(reg.intercept_)
for i, coeff in enumerate(coeffs):
    feature = poly.get_feature_names_out()[selector.get_support()][i]
    formula += " + {:.4f} {}".format(coeff, feature)
print(formula)
print(coeffs)

# Evaluate the model on the training data
y_pred = reg.predict(X_new)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"MSE: {mse:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"R-squared: {r2:.4f}")
```

```python
# Define the formula for v_i
vi_0 = 9.946
def vi(va):
    vi_new = 2 * 11.788 / (0.2482 * np.sqrt((va * 0.748)**2 + (va * 0.664 + vi_0)**2))
    return vi_new

# Define the range for va
va_range = np.linspace(0, 25, num=26)
vi_range = []
# Generate the data points
for va in va_range:
    vi_val = vi(va)
    vi_range.append(vi_val)

# Fit the polynomial
poly_coeff = np.polyfit(va_range, vi_range, deg=3)
poly_func = np.poly1d(poly_coeff)

print(poly_coeff)
print(poly_func)

# Plot the data points and polynomial
plt.plot(va_range, vi_range, 'o', label='Data for v_i, (m = 1.102 kg)')
plt.plot(va_range, poly_func(va_range), label='Polynomial Fit for v_i')
plt.xlabel('Airspeed (v_a) [m/s]')
plt.ylabel('Induced Velocity (v_i) [m/s]')
plt.legend()
plt.show()
```

```python
# Assuming you've already trained your models (lr, poly_reg, grid_search, and theoretical_model)

# Define the intervals for the features
speed_values = np.linspace(0, 20, 100)
weight_values = np.linspace(1000, 1800, 100)
wind_values = np.linspace(0, 5, 100)

# Create a grid of points
speed_grid, weight_grid = np.meshgrid(speed_values, weight_values)
grid_points = np.array([speed_grid.ravel(), weight_grid.ravel()]).T

# Add the wind feature with value 0
wind_values = np.zeros_like(grid_points[:, 0])
grid_points = np.column_stack((grid_points, wind_values))

# Linear regression predictions
lr_pred = np.squeeze(lr.predict(grid_points).reshape(speed_grid.shape))

# Polynomial regression predictions
grid_points_poly = poly.transform(grid_points)

# Use the feature selector to retain only the relevant features
grid_points_selected = selector.transform(grid_points_poly)

# Compute the predictions using the coefficients and the transformed data
poly_pred = reg.intercept_ + np.dot(grid_points_selected, coeffs.T).reshape(speed_grid.shape[0], speed_grid.shape[1])

# XGBoost predictions
xgb_pred = np.squeeze(grid_search.predict(grid_points).reshape(speed_grid.shape))

# Theoretical model predictions (defining it as a function)
def theoretical_model(X):
    m = X[:, 1] / 1000  # weight in kg
    v_a = X[:, 0]  # speed
    g = 9.807  # acceleration due to gravity
    # Calculate the power (P_final) using the given formula
    P_final = 1.25 * m * g * (v_a * 0.5 + (-0.0002131 * v_a**3 + 0.01722 * v_a**2 - 0.5592 * v_a + 9.595))
    return P_final

th_pred = theoretical_model(grid_points).reshape(speed_grid.shape)
```

```python
fig = plt.figure(figsize=(20, 5))
vmin_value = 80
vmax_value = 350

# Theoretical Model plot
ax1 = fig.add_subplot(141, projection='3d')
surf1 = ax1.plot_surface(speed_grid, weight_grid, th_pred , cmap=cm.plasma, alpha=0.9,vmin=vmin_value, vmax=vmax_value)
ax1.set_title('a) Theoretical Model', pad=-80)
ax1.set_xlabel('Speed (m/s)')
ax1.set_ylabel('Weight (g)')
ax1.set_zlabel('Power (W)')

# Linear Regression plot
ax2 = fig.add_subplot(142, projection='3d')
surf2 = ax2.plot_surface(speed_grid, weight_grid,lr_pred , cmap=cm.plasma, alpha=0.9,vmin=vmin_value, vmax=vmax_value)
ax2.set_title('b) Simple Linear Regression', pad=-80)
ax2.set_xlabel('Speed (m/s)')
ax2.set_ylabel('Weight (g)')
ax2.set_zlabel('Power (W)')

# Polynomial Regression plot
ax3 = fig.add_subplot(143, projection='3d')
surf3 = ax3.plot_surface(speed_grid, weight_grid, poly_pred, cmap=cm.plasma, alpha=0.9,vmin=vmin_value, vmax=vmax_value)
ax3.set_title('c) Cubic Polynomial Regression', pad=-80)
ax3.set_xlabel('Speed (m/s)')
ax3.set_ylabel('Weight (g)')
ax3.set_zlabel('Power (W)')

# XGBoost plot
ax4 = fig.add_subplot(144, projection='3d')
surf4 = ax4.plot_surface(speed_grid, weight_grid, xgb_pred, cmap=cm.plasma, alpha=0.9,vmin=vmin_value, vmax=vmax_value)
ax4.set_title('d) XGBoost Regressor', pad=-80)
ax4.set_xlabel('Speed (m/s)')
ax4.set_ylabel('Weight (g)')
ax4.set_zlabel('Power (W)')

# Create a separate subplot for colorbar
ax5 = fig.add_axes([1, 0.25, 0.01, 0.5]) # Adjust the position and size of the colorbar subplot accordingly

# Add colorbar to the rightmost subplot
fig.colorbar(surf4, cax=ax5)

# Plot the actual data points as scatter points
ax1.scatter(X['Speed (m/s)'], X['Weight (g)'], y, c='r', marker='o', s=20, alpha=0.5, label='Actual Data')
ax2.scatter(X['Speed (m/s)'], X['Weight (g)'], y, c='r', marker='o', s=20, alpha=0.5, label='Actual Data')
ax3.scatter(X['Speed (m/s)'], X['Weight (g)'], y, c='r', marker='o', s=20, alpha=0.5, label='Actual Data')
ax4.scatter(X['Speed (m/s)'], X['Weight (g)'], y, c='r', marker='o', s=20, alpha=0.5, label='Actual Data')

# Set the axes limits
ax1.set_xlim(0, 20)
ax1.set_ylim(1000, 2000)
ax1.set_zlim(80, 350)

ax2.set_xlim(0, 20)
ax2.set_ylim(1000, 2000)
ax2.set_zlim(80, 350)

ax3.set_xlim(0, 20)
ax3.set_ylim(1000, 2000)
ax3.set_zlim(80, 350)

ax4.set_xlim(0, 20)
ax4.set_ylim(1000, 2000)
ax4.set_zlim(80, 350)
    
elevation_angle = 15  # Set the elevation angle (in degrees)
azimuth_angle = -120    # Set the azimuth angle (in degrees)

ax1.view_init(elevation_angle, azimuth_angle)
ax2.view_init(elevation_angle, azimuth_angle)
ax3.view_init(elevation_angle, azimuth_angle)
ax4.view_init(elevation_angle, azimuth_angle)

plt.tight_layout()
plt.savefig("models.pdf",bbox_inches='tight')
plt.show()
```
